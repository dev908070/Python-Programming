```python
# name = input("please enter your name: ")
# print("Please Place your order...")
# order = input("Enter number of pizza: ")

order =[]
baselist = []
stylelist = []
toppinglist =[]

pizza ={
    'category':
        {
            'Crush base':{
                'thin':4.00,
                "thick":6.50,
                "woodfired":8.50
            }
    ,
            'Cheese Style':{
                'chedder':5,
                'colby':5,
                'edam':5,
                'emmental':5,
                'gruyere':5,
                'mozzarella':5,
                'provolone' :5,
                'picotta' : 5
            }
            ,
            'topping Choice':
                {
                            'tomato':0.75,
                            'mushrooms': 0.75,
                            'pepperoni':0.50,

                            'onions': 0.25,
                            'black Olives': 0.75,
                            'green Peppers': 0.50,
                            'anchovies': 0.25,
                            'garlic': 0.50,
                            'ham': 1.50,
                            'bacon': 1.25,

            }
        }
}



qty =int(input("How many Pizza You want Order = "))
pizzaqty = qty+1
temp =0
while(True):
    temp=temp+1
    if pizzaqty == temp:
        break
    for values in pizza['category']:
        print("----", values)

    print("Please Select Step 1")


    for values in pizza['category']['Crush base']:
        print("----", values)
    print("Step 1:")
    base = input("Please Select Your Base")
    baselist.append(base)

    for values in pizza['category']['Cheese Style']:
        print("----", values)
    print("Step 2:")
    style = input("Please Select Your Style")
    stylelist.append(style)
    for values in pizza['category']['topping Choice']:
        print("----", values)
    print("Step 3:")
    topping = input("Please Select Your topping Choice :")
    toppinglist.append(topping)

    print(temp,"Your order Selected")


print("No\tBase\tCheeseStyle\tToppingChoice\tPrice")
for v in range(len(baselist)):

    print(v+1,end="\t")
    print(baselist[v],  end="\t")
    print(stylelist[v],  end="\t\t")
    print(toppinglist[v],  end="\t \t\t \t")
    print(pizza['category']['Crush base'][baselist[v]] +pizza['category']['Cheese Style'][stylelist[v]] +pizza['category']['topping Choice'][toppinglist[v]], end=" \t \t\t \t")

    total = pizza['category']['Crush base'][baselist[v]] +pizza['category']['Cheese Style'][stylelist[v]] +pizza['category']['topping Choice'][toppinglist[v]]
    order.append(total)


    print()

odrtotal=0
for  o in order:
    odrtotal=odrtotal+o
print("_"*50)
print("Total:\t\t\t\t\t\t\t\t\t ", end="")
print(odrtotal,end="")

```
